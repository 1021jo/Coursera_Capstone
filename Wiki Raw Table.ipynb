```python
import requests
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page=requests.get(url)
html_string = page.text
soup = BeautifulSoup(html_string, 'lxml') # Parse the HTML as a string
```

```python
table = soup.find_all('table')[0] # Grab the first table
    
df = pd.read_html(str(table))
df=df[0]

# Remove the rows where Borough = Not assigned
df.rename(columns = {0:'Postcode',1:'Borough',2:'Neighbourhood'},inplace=True)
df = df[df.Borough != 'Not assigned']
df
df.drop([0])
```

```python
df = df[df.Borough != 'Not assigned']
df
```

```python
# Combine neighbourhoods with the same postcode

dflist=df.groupby(['Postcode','Borough'])['Neighbourhood'].agg(','.join)
dflist=dflist.to_frame(name = 'Neighbourhood').reset_index()
```

```python
dflist['Neighbourhood'].loc[dflist['Neighbourhood'] == 'Not assigned'] = dflist['Borough']
dflist.shape
```
